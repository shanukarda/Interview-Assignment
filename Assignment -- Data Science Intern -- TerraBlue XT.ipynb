#### 1. Write a program which accepts a sequence of comma separated 4 digit binary numbers as its input and then check whether they are divisible by 5 or not. The numbers that are divisible by 5 are to be printed in a comma separated sequence.
Example:
0100,0011,1010,1001

Then the output should be:
1010

```python
'''
Line 1  : Create a empty list 'num' to store numbers that are divisible by 5.
Line 2  : Take input from user as sequence of comma separated binary numbers and store as list into 'bin_num'
Line 3  : Initialize the for loop on 'bin_num' binary list to check divisiblity by 5.
Line 4  : Convert each binary number into decimal.
Line 5&6: Check if convert number is divisible by 5 then append the binary number into empty list 'num'.
Line 7  : Print binary numbers in 'num'. (Which are divisible by 5).

'''
num = []
bin_num = [x for x in input("Enter comma separater 4 digit binary numbers: ").split(',')]
for i in bin_num:
    a = int(i, 2)
    if a%5 == 0:
        num.append(i)
print("The numbers that are divisible by 5: "+",".join(num))
```

#### 2. A website requires the users to input username and password to register. Write a program to check the validity of password input by users.
Following are the criteria for checking the password:
1. At least 1 letter between [a-z]
2. At least 1 number between [0-9]
1. At least 1 letter between [A-Z]
3. At least 1 character from [$#@]
4. Minimum length of transaction password: 6
5. Maximum length of transaction password: 12
Your program should accept a sequence of comma separated passwords and will check them
according to the above criteria. Passwords that match the criteria are to be printed, each separated by a
comma.

```python
'''
For this program we can use re module which has robust functions to work on regular expressions.
--First we will start a while loop with condition that if user want to enter password for multiple times in case of Invalid
password. 
--Then we will start with second loop to check all password conditions. We will use 'search' function of 're' in if-elif 
conditions to search all required conditions. At the end, in else condition, we will set 'False' of both loops so that
program will stop if password is valid.

--At last after any if condition is false then it will break the second loop control will come to if condition in first
loop. Here we will print "Invalid Password" and ask to user if he want to try again or exit.

'''
import re
c1 = True
while c1:
    password= input("Input your password: ")
    c2 = True
    while c2:  
        if (len(password)<6 or len(password)>12):
            break
        elif not re.search("[a-z]",password):
            break
        elif not re.search("[0-9]",password):
            break
        elif not re.search("[A-Z]",password):
            break
        elif not re.search("[$#@]",password):
            break
        elif re.search("\s",password):
            break
        else:
            print("Valid Password")
            c2 = False
            c1 = False
    if c2:
        print("Not a Valid Password")
        a = input("Enter 'y' to continue or any other key to exit: ")
        if a.lower() == "y":
            c1 = True
        else:
            c1 = False
```

#### 3. You are required to write a program to sort the (name, age, height) tuples by ascending order where name is string, age and height are numbers. The tuples are input by console. The sort criteria is:
1. Sort based on name;
2. Then sort based on age;
3. Then sort by score.
The priority is that name > age > score.

If the following tuples are given as input to the program:

Tom,19,80

John,20,90

Jony,17,91

Jony,17,93

Json,21,85

Then, the output of the program should be:
[('John', '20', '90'), ('Jony', '17', '91'), ('Jony', '17', '93'), ('Json', '21', '85'), ('Tom', '19', '80')]


```python
'''
--We can use 'itemgetter' function from operator module for this type of program.The same program could have been written
with lambda function if we had to sort based on only one item. 'itemgetter' help to fetch multiple values and it return a
tuple as per given values.
--We will start with a empty list 'details' to store all user inputs. It will store tuples.
--We will start a while loop and take input until user will not enter nothing. When user hit enter without entering any
keyword then loop will stopped.
--Using itemgetter, we will sort 'details' list with 'sorted' function and finally we will print sorted list.

'''

from operator import itemgetter
details = []
while True:
    detail = input("name,age,height --->")
    if not detail:
        break
    details.append(tuple(detail.split(',')))
details = sorted(details, key=itemgetter(0,1,2))
print(details)
```

```python
import pandas as pd
data = pd.read_csv("Test_Data.csv",index_col=0)
data.head()
```

1. Extract columns ‘A’,‘B’,’C’,’G’,’R’,’X6’ and ’X8’ for different classes using dictionary and save as a csv for each class.

```python
'''
--First we will declare which columns we want and will create a list of them.
--Create a empty dictionary to store column values of every class.
--Run a for loop on unique values of 'Class' featurn of main data to fetch data for each class.
--In the for loop create another dictionary which will store each column values with key name of column name.
--Run another for loop on each column of each class to fetch values and store into dictionary.
--In outer loop save each columns dictionary into class class dictionary.
--Run another for loop on keys of class dictionary to extract data from class dictionary.
--In for loop create a dataframe for each class and save it as csv using 'to_csv'.
'''

cols = ['A','B','C','G','R','X6','X8']
class_dict = {}
for i in data.Class.unique():
    col_dict = {}
    for col in cols:
        col_dict[col] = list(data[data.Class==i][col])
    class_dict[i] = col_dict
    
for i in class_dict.keys():
    df = pd.DataFrame(class_dict[i])
    df.to_csv(i+"_class.csv")
```

### 5. Create a ROC curve of the model.

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.tree import DecisionTreeClassifier
from itertools import cycle
fig = plt.figure(figsize=(15,10))
X = data_std
y = new_data.iloc[:,-1]


# Binarize the output
y = label_binarize(y, classes=[0, 1, 2, 3, 4, 5])
n_classes = y.shape[1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5, random_state=0)

classifier = OneVsRestClassifier(DecisionTreeClassifier(criterion='entropy',min_samples_split=10,random_state=42))
y_score = classifier.fit(X_train, y_train).predict_proba(X_test)
lw = 1.5

fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
colors = cycle(['blue', 'red', 'green',"orange","magenta","black"])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=lw,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=lw)
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
